```python
# Dictionary of promoted teams per season
promoted_teams = {
    '2014-2015': ['Leicester City', 'QPR', 'Burnley'],
    '2015-2016': ['Norwich City', 'Bournemouth', 'Watford'],
    '2016-2017': ['Hull City', 'Burnley', 'Middlesbrough'],
    '2017-2018': ['Brighton', 'Newcastle United', 'Huddersfield'],
    '2018-2019': ['Fulham', 'Wolves', 'Cardiff City'],
    '2019-2020': ['Norwich City', 'Aston Villa', 'Sheffield United'],
    '2020-2021': ['Fulham', 'West Brom', 'Leeds United'],
    '2021-2022': ['Brentford', 'Watford', 'Norwich City'],
    '2022-2023': ['Fulham', 'Bournemouth', 'Nottingham Forest'],
    '2023-2024': ['Burnley', 'Sheffield United', 'Luton Town']
}
```

```python
premier_ranking_modified = pd.read_excel("/content/Premier_League_Data_Combined_Modified.xlsx")
premier_market_value = pd.read_excel("/content/Premier_Market_Value.xlsx")
premier_goal_keeping = pd.read_excel("/content/Premier_League_Goal_Keeping.xlsx")
```

```python
# Merge premier_market_value into premier_ranking_modified
premier_ranking_modified = pd.merge(premier_ranking_modified, premier_market_value,  how='left', left_on=['Home','Season'], right_on=['Team','Season'])
premier_ranking_modified = premier_ranking_modified.rename(columns={
    'Squad': 'HomeSquad',
    'Age': 'HomeAge',
    'Foreigners': 'HomeForeigners',
    'Total Market Value': 'HomeTotalMarketValue'
})

premier_ranking_modified = pd.merge(premier_ranking_modified, premier_market_value,  how='left', left_on=['Away','Season'], right_on=['Team','Season'])
premier_ranking_modified = premier_ranking_modified.rename(columns={
    'Squad': 'AwaySquad',
    'Age': 'AwayAge',
    'Foreigners': 'AwayForeigners',
    'Total Market Value': 'AwayTotalMarketValue'
})

premier_ranking_modified
```

```python
# Drop the Average Market Value_x Average Market Value_y, Team_x, Team_y columns
premier_ranking_modified = premier_ranking_modified.drop(columns=['Average Market Value_x', 'Average Market Value_y', 'Team_x', 'Team_y'])

premier_ranking_modified
```

```python
# Assuming 'Season' column exists and contains strings like '2023-2024'
premier_ranking_modified['Last Season'] = premier_ranking_modified['Season'].str.slice(0, 4).astype(int) - 1
premier_ranking_modified['Last Season'] = premier_ranking_modified['Last Season'].astype(str) + "-" + (premier_ranking_modified['Last Season'].astype(int) + 1).astype(str)
premier_ranking_modified
```

```python
# Step 1: Create a mapping of seasons to the set of teams that played in each season
teams_by_season = {}

# Get unique seasons from 'Season' and 'Last Season' columns
all_seasons = pd.unique(premier_ranking_modified[['Season', 'Last Season']].values.ravel('K'))

for season in all_seasons:
    # Get matches for the current season
    season_matches = premier_ranking_modified[premier_ranking_modified['Season'] == season]

    # If there are no matches for the season (possible for 'Last Season' that isn't in 'Season'), skip
    if season_matches.empty:
        teams_by_season[season] = set()
        continue

    # Get unique home and away teams
    home_teams = season_matches['Home'].unique()
    away_teams = season_matches['Away'].unique()

    # Combine home and away teams into a set
    teams = set(home_teams).union(set(away_teams))

    # Store the set of teams for the current season
    teams_by_season[season] = teams

# Step 2: Define a function to check if teams were in the last season
def check_team_in_last_season(row):
    home_team = row['Home']
    away_team = row['Away']
    last_season = row['Last Season']

    # Get the set of teams from the last season
    teams_last_season = teams_by_season.get(last_season, set())

    # Check if the home and away teams were in the last season
    home_in_last_season = home_team in teams_last_season
    away_in_last_season = away_team in teams_last_season

    return pd.Series({
        'HomeTeamInLastSeason': home_in_last_season,
        'AwayTeamInLastSeason': away_in_last_season
    })

# Step 3: Apply the function to the DataFrame
premier_ranking_modified[['HomeTeamInLastSeason', 'AwayTeamInLastSeason']] = premier_ranking_modified.apply(
    check_team_in_last_season, axis=1
)

# Now 'premier_match' has two new columns indicating whether home and away teams were in the last season
premier_ranking_modified
```

```python
# Step 2: Initialize a dictionary to store the average save percentage and clean sheet percentage of promoted teams by season
avg_goalkeeping_stats_promoted_by_season = {}

# Get a sorted list of seasons to maintain chronological order
seasons = sorted(promoted_teams.keys())

# Loop through each season to calculate the average save percentage and clean sheet percentage of promoted teams
for idx, season in enumerate(seasons):
    if idx == 0:
        # For the first season, assign NaN since there is no previous season to reference
        avg_goalkeeping_stats_promoted_by_season[season] = {'Save_Percentage': np.nan, 'Clean_Sheet_Percentage': np.nan}
    else:
        last_season = seasons[idx - 1]
        # Get the promoted teams from the last season
        teams = promoted_teams[last_season]

        # Get goalkeeping data for these teams in the last season (their first season in the Premier League)
        team_stats = premier_goal_keeping[
            (premier_goal_keeping['Season'] == last_season) &
            (premier_goal_keeping['Team'].isin(teams))
        ]

        # Calculate the average save percentage and clean sheet percentage
        avg_save_percentage = team_stats['Save_Percentage'].mean()
        avg_clean_sheet_percentage = team_stats['Clean_Sheet_Percentage'].mean()

        avg_goalkeeping_stats_promoted_by_season[season] = {
            'Save_Percentage': avg_save_percentage,
            'Clean_Sheet_Percentage': avg_clean_sheet_percentage
        }

# Step 3: Add the 'Previous_Season' column, but only if the season is not the first one in the dataset
def previous_season(season):
    start_year, end_year = season.split('-')
    prev_start_year = str(int(start_year) - 1)
    prev_end_year = str(int(end_year) - 1)
    return f"{prev_start_year}-{prev_end_year}"

# Adding a check for the first season in the dataset
premier_ranking_modified['Previous_Season'] = premier_ranking_modified['Season'].apply(
    lambda x: previous_season(x) if x != seasons[0] else np.nan
)

# Step 4: Create a function to get the last season stats or assign average for promoted teams
def get_last_season_goalkeeping_stats(team, last_season, current_season, in_last_season):
    if pd.isna(last_season):
        # If there is no previous season, return NaN
        return {'Save_Percentage': np.nan, 'Clean_Sheet_Percentage': np.nan}
    if in_last_season:
        # Team was in the league last season, retrieve actual stats
        last_season_stats = premier_goal_keeping[
            (premier_goal_keeping['Team'] == team) &
            (premier_goal_keeping['Season'] == last_season)
        ]
        if not last_season_stats.empty:
            return {
                'Save_Percentage': last_season_stats.iloc[0]['Save_Percentage'],
                'Clean_Sheet_Percentage': last_season_stats.iloc[0]['Clean_Sheet_Percentage']
            }
        else:
            return {'Save_Percentage': np.nan, 'Clean_Sheet_Percentage': np.nan}
    else:
        # Team was promoted this season, assign average stats of promoted teams from the previous season
        avg_stats = avg_goalkeeping_stats_promoted_by_season.get(current_season, {'Save_Percentage': np.nan, 'Clean_Sheet_Percentage': np.nan})
        return avg_stats

# Step 5: Apply the function to create new columns for Home and Away teams
def apply_goalkeeping_stats(row, is_home):
    team = row['Home'] if is_home else row['Away']
    last_season = row['Previous_Season']
    current_season = row['Season']
    in_last_season = row['HomeTeamInLastSeason'] if is_home else row['AwayTeamInLastSeason']
    stats = get_last_season_goalkeeping_stats(team, last_season,current_season, in_last_season)
    return stats

# Apply for Home teams
premier_ranking_modified['HomeSavePercentage_LastSeason'] = premier_ranking_modified.apply(
    lambda row: apply_goalkeeping_stats(row, is_home=True)['Save_Percentage'], axis=1
)

premier_ranking_modified['HomeCleanSheetPercentage_LastSeason'] = premier_ranking_modified.apply(
    lambda row: apply_goalkeeping_stats(row, is_home=True)['Clean_Sheet_Percentage'], axis=1
)

# Apply for Away teams
premier_ranking_modified['AwaySavePercentage_LastSeason'] = premier_ranking_modified.apply(
    lambda row: apply_goalkeeping_stats(row, is_home=False)['Save_Percentage'], axis=1
)

premier_ranking_modified['AwayCleanSheetPercentage_LastSeason'] = premier_ranking_modified.apply(
    lambda row: apply_goalkeeping_stats(row, is_home=False)['Clean_Sheet_Percentage'], axis=1
)

```

```python
# Create a new DataFrame to store the rolling statistics
home_team_stats = []

# Loop through each row in the dataset
for index, row in premier_ranking_modified.iterrows():
    home_team = row['Home']
    away_team = row['Away']
    match_date = row['Date']

    # Filter to get the matches between the home and away team that happened before the current match date
    past_matches = premier_ranking_modified[
        (
            ((premier_ranking_modified['Home'] == home_team) & (premier_ranking_modified['Away'] == away_team)) |
            ((premier_ranking_modified['Home'] == away_team) & (premier_ranking_modified['Away'] == home_team))
        ) &
        (premier_ranking_modified['Date'] < match_date)
    ].sort_values(by='Date', ascending=False)

    # Get the last 3 matches between these teams
    last_three_matches = past_matches.head(3)

    # Calculate "Goals For" and "Goals Against" for the home team
    home_goals_for = 0
    home_goals_against = 0

    for past_match in last_three_matches.itertuples():
        if past_match.Home == home_team:
            # Replace any en dash or similar characters with a hyphen before splitting
            score_parts = past_match.Score.replace('–', '-').replace('—', '-').split('-')
            home_goals_for += int(score_parts[0])
            home_goals_against += int(score_parts[1])
        else:
            # Replace any en dash or similar characters with a hyphen before splitting
            score_parts = past_match.Score.replace('–', '-').replace('—', '-').split('-')
            home_goals_for += int(score_parts[1])
            home_goals_against += int(score_parts[0])

    # Append the calculated goals for and against to the new DataFrame
    home_team_stats.append({
        'Home': home_team,
        'Away': away_team,
        'Date': match_date,
        'Last_Three_Home_Goals_For': home_goals_for,
        'Last_Three_Home_Goals_Against': home_goals_against
    })

# Convert the list of dictionaries to a DataFrame and merge it with the original DataFrame
home_team_stats_df = pd.DataFrame(home_team_stats)
premier_ranking_modified = pd.merge(premier_ranking_modified, home_team_stats_df, on=['Home', 'Away', 'Date'], how='left')

# Display the modified DataFrame
premier_ranking_modified
```

```python
# Assuming 'premier_ranking_modified' is already loaded with your dataset
# Step 1: Convert 'Date' to datetime
premier_ranking_modified['Date'] = pd.to_datetime(premier_ranking_modified['Date'])

# Step 2: Create a DataFrame to store points data for each matchup
home_team_points_stats = []

# Loop through each row in the dataset
for index, row in premier_ranking_modified.iterrows():
    home_team = row['Home']
    away_team = row['Away']
    match_date = row['Date']

    # Filter to get all previous matches between the two teams before the current match date
    past_matches = premier_ranking_modified[
        (
            ((premier_ranking_modified['Home'] == home_team) & (premier_ranking_modified['Away'] == away_team)) |
            ((premier_ranking_modified['Home'] == away_team) & (premier_ranking_modified['Away'] == home_team))
        ) &
        (premier_ranking_modified['Date'] < match_date)
    ].sort_values(by='Date', ascending=False)

    # Get the last 3 matches between these teams
    last_three_matches = past_matches.head(3)

    # Calculate the points earned by the home team in those matches
    points_earned = 0
    for past_match in last_three_matches.itertuples():
        # Access the 'Game_Outcome' column using index from the named tuple
        if past_match.Home == home_team and past_match[premier_ranking_modified.columns.get_loc('Game Outcome') + 1] == 'Win':
            points_earned += 3
        elif past_match.Away == home_team and past_match[premier_ranking_modified.columns.get_loc('Game Outcome') + 1] == 'Loss':
            points_earned += 3
        elif past_match[premier_ranking_modified.columns.get_loc('Game Outcome') + 1] == 'Draw':
            points_earned += 1
        # Loss yields 0 points, so no need to add

    # Append the calculated points to the list
    home_team_points_stats.append({
        'Home': home_team,
        'Away': away_team,
        'Date': match_date,
        'Last_Three_Home_Points': points_earned
    })

# Convert the list of dictionaries to a DataFrame and merge it with the original DataFrame
home_team_points_stats_df = pd.DataFrame(home_team_points_stats)
premier_ranking_modified = pd.merge(premier_ranking_modified, home_team_points_stats_df, on=['Home', 'Away', 'Date'], how='left')
```
